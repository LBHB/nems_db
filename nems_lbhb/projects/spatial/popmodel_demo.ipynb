# Example whole-site LN_Pop fit

Find all sites for this batch

```python
batch = 338
siteids, cellids = db.get_batch_sites(338)
```

### Define the model architecture and fit parameters

Pick a siteid

```python
siteid='CLT011a'
cellid = [c for c in cellids if c.startswith(siteid)]
siteid, cellid
```

Define architecture and fit parameters

```python
load_kw = 'gtgram.fs100.ch18.bin6-ld-norm.l1-sev'
fit_kw = 'lite.tf.init.lr1e3.t3.es20.rb3-lite.tf.lr1e4.t5e4'
#fit_kw = 'lite.tf.init.lr1e3.t3.es20'

modelname = f"{load_kw}_LNpop.10xNxRx30_{fit_kw}"
```

### Fit but don't save

```python
xf,ctx = xform_helper.fit_model_xform(cellid=siteid, batch=batch, modelname=modelname, returnModel=True, saveInDB=False)
```

### Fit and save

```python
savefile = xform_helper.fit_model_xform(cellid=siteid, batch=batch, modelname=modelname, saveInDB=True)
```
